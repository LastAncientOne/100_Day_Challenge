```python
# Day 29 - Calculate Beta
from scipy import stats

def stock_beta():
    stock = 'AMD'
    market = '^GSPC'
    start = '2017-01-01' 
    end = '2018-01-01'
    mk = pdr.get_data_yahoo(market, start, end)
    df = pdr.get_data_yahoo(stock, start, end)
    
    # joining the closing prices of the two datasets 
    monthly_prices = pd.concat([df['Adj Close'], mk['Adj Close']], axis=1)
    monthly_prices.columns = ['Stock', '^GSPC']

    # calculate monthly returns
    monthly_returns = monthly_prices.pct_change(1)
    clean_monthly_returns = monthly_returns.dropna(axis=0)  # drop first missing row


    # split dependent and independent variable
    X = clean_monthly_returns['^GSPC']
    y = clean_monthly_returns['Stock']
    
    beta, intercept, r_value, p_value, std_err = stats.linregress(X, y)
    print("beta:", beta)
    return

stock_beta()
```

```python
# Day 35 - Cost of Equity = Risk-Free Rate of Return + Beta * (Market Rate of Return - Risk-Free Rate of Return)
from scipy import stats

def Cost_Of_Equity():
    ret = df['Adj Close'].pct_change(1).dropna()
    mrk_ret = mk['Adj Close'].pct_change(1).dropna()
    Rf = 0.01
    mrk_rate_ret = (mk['Adj Close'][-1] - mk['Adj Close'][0])/ mk['Adj Close'][0]
    (beta, alpha, r_value, p_value, std_err)=stats.linregress(ret, mrk_ret)
    COE = Rf + beta * (mrk_rate_ret - Rf)
    return COE

Cost_Of_Equity()    
```

```python
# Day 36 - Roll's model(1984)
from math import sqrt

def Roll_model():
    d = np.diff(df['Adj Close'])
    cov_ = np.cov(d[:-1], d[1:])
    if cov_[0,1] < 0:
        print("Roll's model spread for ", stock, "is", round(2*sqrt(-cov_[0,1]), 3))
    else:
        print("Covariance is positive for ", stock, "positive", round(cov_[0,1], 3))
    return

Roll_model()
```

```python
# Day 37 - Amihud's model(2002)
def Amihud_model():
    p = np.array(df["Adj Close"])
    dollar_Volume = np.array(df["Volume"]*p)
    ret = np.array((p[1:] - p[:1]) / p[1:])
    illiquidity = np.mean(np.divide(abs(ret), dollar_Volume[1:]))
    return illiquidity

Amihud_model()
```

```python
# Day 38 - Durbin-Watson
from scipy import stats
import statsmodels.formula.api as sm
import statsmodels.stats.stattools as tools

def Durbin_Watson():
    ret = df['Adj Close'].pct_change(1).dropna()
    mk_ret = mk['Adj Close'].pct_change(1).dropna()
    New_Data = pd.DataFrame({"Y":ret, "X": mk_ret})
    result = sm.ols(formula="Y ~X", data=New_Data).fit()
    print(result.params)
    residuals=result.resid
    print("Durbin Watson")
    print(tools.durbin_watson(residuals))
    return

Durbin_Watson()
```

```python
# Day 39 - Value-at-Risk
from scipy.stats import norm
from math import sqrt

def VaR():
    ret = df['Adj Close'].pct_change(1).dropna()
    shares = 100
    confidence_level = 0.99
    days = 10
    z = norm.ppf(confidence_level)
    position = shares*df['Adj Close'][0]
    VaR = position*z*np.std(ret)*sqrt(days)
    print("Holding:", position, " VaR:", round(VaR, 4), "in", days, "Days")
    return

VaR()
```

```python
# Day 40 - Beta(Another way to calculate)
def beta():
    ret = df['Adj Close'].pct_change(1).dropna()
    mk_ret = mk['Adj Close'].pct_change(1).dropna()
    # matrix of returns, market
    m = np.matrix([ret, mk_ret])
    # cov of m  / std dev of market returns
    beta = np.cov(m)[0][1] / np.std(mk_ret)
    return beta
    
beta()
```

```python
# Day 41 - Alpha (Another way to calculate)
def alpha():
    mk_X = mk['Adj Close'].pct_change(1).dropna()
    ret_Y = df['Adj Close'].pct_change(1).dropna()
    m = np.matrix([ret_Y, mk_X])
    beta = np.cov(m)[0][1] / np.std(mk_X)
    alpha = np.mean(ret_Y) - beta * np.mean(mk_X)
    return alpha

alpha()
```

```python
# Day 42 - Value-At-Risk
def VaR():
    ret = df['Adj Close'].pct_change(1).dropna()
    mk_ret = mk['Adj Close'].pct_change(1).dropna()
    m = np.matrix([ret, mk_ret])
    beta = np.cov(m)[0][1] / np.std(mk_ret)
    alpha = np.mean(ret) - beta * np.mean(mk_ret)
    # calculates the historical simulation var of the returns
    sorted_returns = np.sort(ret)
    # calc index associated with alpha
    index = int(alpha * len(sorted_returns))
    # var should be positive
    return abs(sorted_returns[index])

VaR()
```

```python
# Day 43 - Conditional value-at-risk
def CVaR():
    ret = df['Adj Close'].pct_change(1).dropna()
    mk_ret = mk['Adj Close'].pct_change(1).dropna()
    m = np.matrix([ret, mk_ret])
    beta = np.cov(m)[0][1] / np.std(mk_ret)
    alpha = np.mean(ret) - beta * np.mean(mk_ret)
    # calculates the condition VaR of the returns
    sorted_returns = np.sort(ret)
    # calculate the index associated with alpha
    index = int(alpha * len(sorted_returns))
    # calculate total var beyond alpha
    sum_var = sorted_returns[0]
    for i in range(1, index):
        sum_var += sorted_returns[i]
    
    # return average VaR
    # CVaR should be positive
    return abs(sum_var/index)

CVaR()
```

```python
# Day 49 - Modigliani Ratio 
def modigliani_ratio():
    ret = df['Adj Close'].pct_change(1).dropna()
    mk_ret = mk['Adj Close'].pct_change(1).dropna()
    rf = 0.01
    mrk_rate_ret = (mk['Adj Close'][-1] - mk['Adj Close'][0])/ mk['Adj Close'][0]
    m = np.matrix([ret, mk_ret])
    beta = np.cov(m)[0][1] / np.std(mk_ret)
    er = rf + beta*(mrk_rate_ret-rf)
    np_rf = np.empty(len(ret))
    np_rf.fill(rf)
    rdiff = ret - np_rf
    bdiff = mk_ret - np_rf
    mr = (er - rf) * np.std(rdiff) / np.std(bdiff) + rf
    return mr

modigliani_ratio()
```

```python
# Day 51 - Lower Partial Moments
def lower_partial_moments():
    ret = df['Adj Close'].pct_change(1).dropna()
    ret = np.array(ret)
    threshold = 0.0
    order = 1
    # This method returns a lower partial moment of the returns
    # Create an array he same length as returns containing the minimum return threshold
    threshold_array = np.empty(len(ret))
    threshold_array.fill(threshold)
    # Calculate the difference between the threshold and the returns
    diff = threshold_array - ret
    # Set the minimum of each to 0
    diff = diff.clip(min=0)
    # Return the sum of the different to the power of order
    lpm = np.sum(diff ** order) / len(ret)
    return lpm
    
lower_partial_moments()
```

```python
# Day 54 - Drawdowns in Adj Closing Prices
def drawdown():
    maxi = 0
    mini = 0
    for i in range(len(df['Adj Close']))[1:]:
        maxj = 0
        max = 0
        for j in range(i+1, len(df['Adj Close'])):
            if df['Adj Close'][i] - df['Adj Close'][j] > max:
                maxj = j
                max = df['Adj Close'][i] - df['Adj Close'][j]
        if max > df['Adj Close'][maxi] - df['Adj Close'][mini]:
            maxi = i
            mini = maxj
    return (df['Adj Close'][maxi], df['Adj Close'][mini])

drawdown()
```

```python
# Day 89 - Support and Resistance
from sklearn.svm import SVR

def Sup_Res():
    data = None
    volumes = list(df['Volume'])
    average = sum(volumes)/len(volumes)
    
    if data==None or data==1:
        pertinent = [volumes.index(x) for x  in volumes if x>= average]
    else:
        pertinent = [volumes.index(x) for x  in volumes if x>= average][:data]
    
    
    prices = list(df['Adj Close'])
    #ignore the low volume price
    prices_pert = [prices[i] for i in pertinent[::-1]]
    svr_lin = SVR(kernel= 'linear', C= 1e3) 
    X=np.arange(1,len(prices_pert)+1,1.0)
    X=np.reshape(X,(len(X),1))
    y=prices_pert
    y_lin = svr_lin.fit(X, y).predict(X)
    diff = [(y_lin[x]-y[x])/y_lin[x] for x in range(len(y))]
    #support coefficient
    
    bpd = max(diff)
    #resistance coefficient
    
    spd = min(diff)
    volatility = bpd-spd
    print("volatiliy: %.3f" %(volatility))
    y_supp = [ i*(1-bpd) for i in y_lin ]
    y_res = [ i*(1+abs(spd)) for i in y_lin ]
    
 
    plt.title(stock)
    plt.plot(y_lin)
    plt.plot(y)
    plt.plot(y_supp,label="Support")
    plt.plot(y_res,label="resistance")
    plt.legend()
    plt.show()
    buying_point = diff.index(max(diff))
    gain = y[-1]/y[buying_point]
    return len(pertinent)

Sup_Res()
```

```python
# Day 91 - Trend Lines
def fit_line(t, y):
    A = np.vstack([t, np.ones_like(t)]).T
    return np.linalg.lstsq(A, y)[0]

def trend_lines():
    h = df['High']
    l = df['Low']
    c = df['Adj Close']
    pivots = (h + l + c) / 3
    t = np.arange(len(c))
    sa, sb = fit_line(t, pivots - (h - l))
    ra, rb = fit_line(t, pivots + (h - l))
    support = sa * t + sb
    resistance = ra * t + rb
    plt.figure(figsize=(16,10))
    plt.plot(t, c)
    plt.plot(t, support)
    plt.plot(t, resistance)
    plt.show()
    print("Tomorrows support:", sa * (t[-1] + 1) + sb)
    print("Tomorrows resistance:", ra * (t[-1] + 1) + rb)
    return

trend_lines()
```

```python
# Day 99 - Stock for Geometric Brownian Motion
from numpy.random import standard_normal

def GBM():
    T = 1                              
    mu = np.mean(ret)*252
    sigma = np.std(ret)*np.sqrt(252)
    S0 = df['Adj Close'][0]
    N = 2**6
    dt = 1/N    
    t = np.linspace(0,T,N)
    W = np.random.standard_normal(size = N) 
    W = np.cumsum(W)*np.sqrt(dt)
    X = np.add(np.dot((mu-0.5*sigma**2),t) , sigma*W)
    S = S0*np.exp(X)
    plt.plot(t, S)
    plt.ylabel('Stock Price')
    plt.title('Geometric Brownian Motion')
    plt.show()
    return

GBM()
```
